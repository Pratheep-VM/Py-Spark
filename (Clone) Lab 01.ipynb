withColumn[] and withColumnRenamed[]

```python
df=df.withColumnRenamed("Salary","SalaryUSD")
df.show()
df=df.withColumn("Bonus",df.SalaryUSD*0.9)
df.show()
```

**`distinct()`**: Removes duplicate rows from the DataFrame to ensure data integrity.
**`dropDuplicates()`**: Used to ensure data uniqueness. Unlike distinct(), this function allows us to deduplicate data based on specific columns (e.g., ensuring only one record exists per Transaction ID), ignoring differences in other non-key columns.

```python
df.distinct().show()
df.dropDuplicates(["ID"]).show()
```

```python
df.groupBy("Age").count().show()
df.groupBy("Age").sum("SalaryUSD").show()
```

**filter()**: This transformation restricts the DataFrame to return only rows that satisfy the specified condition

```python
df.filter(df.Age>30).show()

```

```python
df.sort("SalaryUSD").show()
df.sort(df.SalaryUSD.desc()).show()
df.orderBy("SalaryUSD").show()
```
